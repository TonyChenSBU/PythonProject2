## 2. Implementing Gradient Descent Algorithm (60 pts)

Given a function $f(x) = \sqrt{x^2+5}$, $x\in \mathbf{R}$.

Fact: This function is L-smooth.
0. Find the gradient of this function and the minimum value of this function analytically.

Use the last digit of your SB ID divided by 5 and then plus 1 as the initial guess (for example, if your ID ends with 9, your initial guess $x_0$ will be $9\div5 + 1= 2.8$).
3. Record the values of $x_k$ at the $k$-th iteration during GD and report $x_T$.

5. For each of the step sizes 5, 3, 1, and 0.5, perform gradient descent and record the values of $x_k$ in each step $k$. Plot $f(x_{k-1}) - f(x_{k})$ v.s. $k$ for each step size.
```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = sqrt(x^2 + 5)
x = np.linspace(-5, 5, 1000)
y = np.sqrt(x**2 + 5)

# Find the minimum value of the function
min_point_y = min(y)
min_point_x = np.sqrt(min_point_y**2  - 5)

# Plot the function f(x), np.linspace might be helpful.
plt.plot(x, y, label='f(x) = sqrt(x^2 + 5)')
plt.scatter(min_point_x, min_point_y, color='red', label='Minimum Point ')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = sqrt(x^2 + 5)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Define the derivative of the function f(x), for gradient descent
# f'(x) = x / np.sqrt(x**2 + 5)
# Written within gradient_descent function

# write a function to run the Gradient Descent Algorithm, take input arguments x_0, step_size, and num_iterations; return a list/array of x_k values.
def gradient_descent(x_0, step_size, num_iterations):
    x_values = [x_0]
    for _ in range(num_iterations):
        x_0 = x_0 - step_size * (x_0 / np.sqrt(x_0**2 + 5))
        x_values.append(x_0)
    return x_values

# set up/assign initial values
x_0 = 2.8
step_size = 1
num_iterations = 50


# Run gradient descent by calling the function
x_values = gradient_descent(x_0, step_size, num_iterations)

# report x_T by printing out the value
x_T = x_values[50]
print(f"The value of x_T = {x_T}")

# Initialize an empty list to store f_values
f_values = []

# Iterate through x_values
for x_val in x_values:
    # Calculate f(x) and append it to f_values
    f_val = np.sqrt(x_val**2 + 5)
    f_values.append(f_val)

# Plot the f(x_k) values vs. iteration number
plt.plot(range(num_iterations+1), f_values)
plt.xlabel('Iteration')
plt.ylabel('f(x_k)')
plt.title('f(x_k) vs. Iteration')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# create a list containing step sizes
step_sizes = [5, 3, 1, 0.5]

# initialize a list to store the results from different step sizes
results = []

# loop through the step sizes and perform GD and record results
for step_size in step_sizes:
    x_values = gradient_descent(x_0, step_size, num_iterations)

    # Initialize an empty list to store f_values
    f_values = []
    for x_val in x_values:
        # Calculate f(x) and append it to f_values
        f_val = np.sqrt(x_val**2 + 5)
        f_values.append(f_val)

    # Initialize an empty list to store differences
    differences = []
    for i in range(1, len(f_values)):
        # Calculate f(x_{k-1}) - f(x_k) and append it to differences
        diff = f_values[i-1] - f_values[i]
        differences.append(diff)

    # Append the results to the results list
    results.append((x_values, differences))

# plot multiple curves in a single graph
for i, (x_values, differences) in enumerate(results):
    plt.plot(range(1, num_iterations+1), differences, label=f'Step Size: {step_sizes[i]}')

plt.xlabel('Iteration')
plt.ylabel('f(x_{k-1}) - f(x_k)')
plt.title('f(x_{k-1}) - f(x_k) vs. Iteration')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(0)   # change the seed to be the last two digits of your SB ID.
X = 2 * np.random.rand(100, 1)  # Random feature values between 0 and 2
y = 3 * X + 1 + np.random.randn(100, 1)  # Linear relationship with noise

# Add a column of ones to X for the bias terms (intercept) in the normal equation



# Find theta using the normal equation



# Print the intercept and coefficients of the regression plane


# Plot the generated data and the regression plane
"""

import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(89) # change the seed to be the last two digits of your SB ID.
X = 2 * np.random.rand(100, 2)  # Random feature values between 0 and 2
y = 4 + 4 * X[:, 0] + 0.5 * X[:, 1] + np.random.randn(100) # Linear relationship with noise

# Add a column of ones to X for the bias terms (intercept) in the normal equation
X_b = np.c_[np.ones((100, 1)), X]

# Find theta using the normal equation
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# Extract coefficients
intercept = theta[0]  # This is the intercept term (the starting point of the plane)
coef1 = theta[1]      # This is the coefficient for the first feature (X1)
coef2 = theta[2]      # This is the coefficient for the second feature (X2)

# From https://www.geeksforgeeks.org/3d-surface-plotting-in-python-using-matplotlib/
# Generate grid points
x_range = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
y_range = np.linspace(min(X[:, 1]), max(X[:, 1]), 100)
X1_grid, X2_grid = np.meshgrid(x_range, y_range)

# Calculate predicted values for the grid
z_grid = intercept + coef1 * X1_grid + coef2 * X2_grid

# Plot the generated data and the regression plane
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the original data points in red
ax.scatter(X[:, 0], X[:, 1], y, c='b', label='Original Data', marker='o')

# Surface plot
surface = ax.plot_surface(X1_grid, X2_grid, z_grid, cmap='viridis', alpha=0.8)

# Add a colorbar
fig.colorbar(surface)

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

# Add title
ax.set_title('Linear Regression Plane')

# Add legend
ax.legend()
plt.show()
```

## 4. Logistic Regression with Numpy and Graphing with Matplotlib (20 pts)

In class, we implemented a logistic regression model to perform binary classification. Here, I provided the exact same code that we discussed in class, but the values for learning_rate and num_epochs are modified. I also removed all the comments. You task is to
1. Tune the Learning Rate and Number of Epochs: Your first task is to fine-tune the values for the learning_rate and num_epochs parameters. Your goal is to identify suitable values that will enable us to converge to a set of parameter values $\theta$ closely approximating the optimal parameter values $\theta^*$.
2. Code Explanation: To enhance code comprehension, please augment the code with meaningful comments. These comments should elucidate the purpose and functionality of each code segment, making it easier for readers to understand the logistic regression implementation.

By accomplishing these tasks, we aim to achieve a better understanding of the logistic regression model's behavior and its parameter optimization process.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function for logistic regression
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Setting random seed for reproducibility
np.random.seed(595)

# Generating random data
X = np.random.rand(2, 100)
w_true = np.array([1.5, -2.5])
b_true = 1.0  # True bias
probabilities = sigmoid(np.dot(w_true.T, X) + b_true)
Y = (probabilities > 0.5).astype(int)
X = X + 0.3 * np.random.rand(2, 100) - 0.1 * np.random.rand(2, 100)

# Splitting data into train and test sets
X_train, X_test = X[:,:80], X[:,80:]
Y_train, Y_test = Y[:80], Y[80:]

# Initializing parameters
w = np.zeros(X.shape[0])
b = 0.0

# Fine-tuning learning rate and number of epochs
learning_rate = .006 # Adjusting this value based on trial and error
num_epochs = 10000 # Adjusting this value based on trial and error

# Training the model
for epoch in range(num_epochs):
    A_train = sigmoid(np.dot(w.T, X_train) + b)

    dJdw = np.dot(X_train, (A_train - Y_train).T) / len(Y_train)
    dJdb = np.mean(A_train - Y_train)

    w -= learning_rate * dJdw
    b -= learning_rate * dJdb

# Predicting on training set
A_train = sigmoid(np.dot(w.T, X_train) + b)
predictions_train = (A_train > 0.5).astype(int)

# Predicting on test set
A_test = sigmoid(np.dot(w.T, X_test) + b)
predictions_test = (A_test > 0.5).astype(int)

# Calculating accuracy
train_accuracy = np.mean(predictions_train == Y_train)

test_accuracy = np.mean(predictions_test == Y_test)

# Printing results
# Visualizing the decision boundary
print(f"Training Set Accuracy: {train_accuracy:.2f}")
print(f"Test Set Accuracy: {test_accuracy:.2f}")
```

```python
# Scatter plot of the first 80 data points in feature space,
# with colors representing class labels (0 or 1)
plt.scatter(X[0, :80], X[1,:80], c=Y[:80], cmap=plt.cm.Paired)

# Set the x-axis limits from -0.2 to 1.2
plt.xlim(-0.2, 1.2)

# Set the y-axis limits from -0.2 to 1.2
plt.ylim(-0.2, 1.2)

# Get the current axes
ax = plt.gca()

# Get the x,y-axis limits
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create a grid of evenly spaced points in the x and y dimensions
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))

# Compute the linear combination of weights and input features
# plus the bias term for each point on the grid
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b

# Apply the sigmoid function to the linear combination
Z = sigmoid(Z)

# Reshape Z to match the shape of xx and yy
Z = Z.reshape(xx.shape)

# Contour plot of the decision boundary at the level 0.5,
# with dashed lines and semi-transparent fill
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])

# Plot
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')
plt.show()
```
